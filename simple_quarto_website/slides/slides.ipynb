````python
#| echo: true
#| code-fold: true
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("penguins")
sns.set_theme(style = 'white', palette = 'Pastel1')
FS = 18
g = sns.jointplot(data = df, x = "flipper_length_mm", y= "bill_length_mm", hue = "species", s = 150)
g.ax_joint.set_xlabel("Penguin flipper length (mm)", fontsize = FS)
g.ax_joint.set_ylabel("Penguin bill length (mm)", fontsize = FS)
g.ax_joint.tick_params(axis = "both", which = "major", labelsize = FS)
plt.savefig('plot_2.pdf', bbox_inches = "tight")
````
